### Работа с пропущенными значениями

Пропущенных значений довольно много. Как бы я поступила с каждым из столбцов:

* **ceiling_height** - на мой взгляд, высота потолков скорее всего одинакова в домах, которые расположены рядом, потому что они скорее всего одного типа постройки. Поэтому заполним пропуски, исходя из высоты потолков домов, которые находятся рядом.
* **floors_total** - данный столбец можем заполнить по той же логике, что и высоту потолков.
* **living_area** - здесь также будем отталкиваться от логики, что если дома расположены рядом, то скорее всего и площадь у них может совпадать. Однако нужно сделать проверку по соотношению с общей площадью и количеством комнат.
* **is_apartment** - очень много пропущенных значений, удалять нельзя. Посмотрим на распределение значений здесь, вероятнее всего, здесь пропуски, так как квартира не является апартаментами, поэтому это значение при заполнении просто пропускали.
* **kitchen_area** - пропуски заполним так же, как и для жилой площади.
* **balcony** - видимо, здесь балконов нет, поэтому заменим на 0.
* **locality_name** - данных слишком мало, я думаю, что мы можем ими пожертвовать. Название населенного пункта нам понадобится, логически это значение мы заполнить не можем, поэтому просто удалим.
* **airports_nearest** - значений много, удалять их не стоит, возможно, эти значения не указаны, так как ближайший аэропорт вне зоны досягаемости.
* **cityCenters_nearest** - количество пропусков очень похоже на количество пропусков с аэропортами. Может быть эта недвижимость находится не в городе, а в деревне? Поэтому считать эти показатели довольно сложно, и они были бы выбросами. Думаю, тут мы пропуски можем оставить.
* **parks_around3000** - здесь мы заполним значения 0, так как, видимо, в раудисе 3км, парков нет.
* **ponds_around3000** - здесь мы заполним значения 0, так как, видимо, в раудисе 3км, водоемов нет.
* **parks_nearest** - заполним по аналогии с **parks_around3000**.
* **ponds_nearest** - заполним по аналогии с **ponds_around3000**.
* **days_exposition** - посмотрим на среднее и медианное значение, а также на распределение количества пропусков, и примем решение.

```python
hf = estate_data[estate_data['cityCenters_nearest'].notna() & estate_data['airports_nearest'].notna()].groupby(['cityCenters_nearest', 'airports_nearest'])

print('Количество пропусков:', estate_data['ceiling_height'].isna().sum())

estate_data['ceiling_height'] = estate_data['ceiling_height'].fillna(hf['ceiling_height'].transform(lambda group: (group.median())))

print('Количество пропусков после преобразования:', estate_data['ceiling_height'].isna().sum())
```

```python
# оставшиеся пропуски можем заменить медианным значением

estate_data['ceiling_height'] = estate_data['ceiling_height'].fillna(estate_data['ceiling_height'].median())
estate_data['ceiling_height'].isna().sum()
```

```python
# пропуски по количеству этажей можем попробовать заполнить по той же логике, что и высоту потолков

print('Количество пропусков:', estate_data['floors_total'].isna().sum())

estate_data['floors_total'] = estate_data['floors_total'].fillna(hf['floors_total'].transform(lambda group: (group.median())))

print('Количество пропусков после преобразования:', estate_data['floors_total'].isna().sum())
```

```python
# оставшиеся пропуски можем заменить медианным значением

estate_data['floors_total'] = estate_data['floors_total'].fillna(estate_data['floors_total'].median())
estate_data['floors_total'].isna().sum()
```

```python
# заполним часть пропусков исходя из соотношения общей площади и количества комнат

print('Количество пропусков:', estate_data['living_area'].isna().sum())

gb = estate_data.groupby(['total_area','rooms'])
estate_data['living_area'] = estate_data['living_area'].fillna(gb['living_area'].transform(lambda group: group.mean() / (group.name[0])) * estate_data['total_area'])

print('Количество пропусков после преобразования:', estate_data['living_area'].isna().sum())
```

```python
# теперь оставшиеся пропуски заполним, исходя из домов поблизости

estate_data['living_area'] = estate_data['living_area'].fillna(hf['living_area'].transform(lambda group: (group.median())))

print('Количество пропусков после преобразования:', estate_data['living_area'].isna().sum())
```

```python
estate_data['living_area'] = estate_data['living_area'].fillna(estate_data['living_area'].mean())
estate_data['living_area'].isna().sum()
```

```python
# с кухней поступим так же как и с жилой площадью

print('Количество пропусков:', estate_data['kitchen_area'].isna().sum())

estate_data['kitchen_area'] = estate_data['kitchen_area'].fillna(gb['kitchen_area'].transform(lambda group: group.mean() / (group.name[0])) * estate_data['total_area'])

print('Количество пропусков после преобразования:', estate_data['kitchen_area'].isna().sum())
```

```python
# теперь оставшиеся пропуски заполним, исходя из домов поблизости

estate_data['kitchen_area'] = estate_data['kitchen_area'].fillna(hf['kitchen_area'].transform(lambda group: (group.median())))

print('Количество пропусков после преобразования:', estate_data['kitchen_area'].isna().sum())
```

```python
estate_data['kitchen_area'] = estate_data['kitchen_area'].fillna(estate_data['kitchen_area'].mean())
estate_data['kitchen_area'].isna().sum()
```

```python
# посмотрим на распределение значений столбца is_apartment

estate_data['is_apartment'].value_counts()
```

### Проведение необходимых расчетов и добавление столбцов

В данном пункте мы займемся расчетами, что нам нужно рассчитать согласно условию:

* цену квадратного метра;
* 		день недели, месяц и год публикации объявления;
* 		этаж квартиры; варианты — первый, последний, другой;
* 		соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
#начнем по порядку - с расчета стоимости за кв.м. Для этого нам нужно будет создать 
#новый столбец, назовем его price_for_m

estate_data['price_for_m'] = estate_data['last_price'] / estate_data['total_area']
estate_data['price_for_m'] = estate_data['price_for_m'].astype('int')
estate_data['price_for_m'].head(5)
```

```python
#сделаем три отдельных столбца для дня недели, месяца и года, 
#назовем соответственно - week_day_exp, month_exp, year_exp

estate_data['weekday_exp'] = estate_data['first_day_exposition'].dt.weekday
estate_data['month_exp'] = estate_data['first_day_exposition'].dt.month
estate_data['year_exp'] = estate_data['first_day_exposition'].dt.year
estate_data.head(5)
```

```python
#для этажей квартир сделаем категоризацию, написав соответствующую функцию


def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'


estate_data['floor_group'] = estate_data.apply(floor_group, axis=1)

estate_data['floor_group'].value_counts()
```

```python
#создадим два столбца, которые будут содержать данные об отношении жилой площади к общей, 
#и площади кухни к общей. Назовем соответственно living_vs_total и kitchen_vs_total

estate_data['living_vs_total'] = estate_data['living_area'] / estate_data['total_area']
estate_data['kitchen_vs_total'] = estate_data['kitchen_area'] / estate_data['total_area']
estate_data.head(5)
```

<br>Чтобы удобнее было анализировать время продажи квартир я хочу вычислить последний день продажи (дата), чтобы потом можно было посмотреть продажи по дням недели и сезонности.

```python
estate_data['last_day_sale'] = estate_data.apply(lambda row: row['first_day_exposition'] + pd.Timedelta(days=row['days_exposition']), axis=1)
estate_data.head(5)
```

Сделали столбец с днем продажи, теперь сделаем категоризацию по цене. Так как разброс цен у нас большой, я буду делать большие группы, разбивку по сегментам сделаю такую:

* "**Другое**" - 12190 <= x < 1700000 - эти значения слишком маленькие по стоимости и их довольно немного в выборке (менее 1000 шт), поэтому предлагаю сделать такой большой срез.
* "**Низший**" - 1700000 <= x < 3400000 (до первого квартиля)
* "**Ниже среднего**" - 3400000 <= x < 4650000 (до медианы)
* "**Средний**" - 4650000 <= x <= 6799000 (до второго квартиля)
* "**Выше среднего**" - 7000000 <= x < 30000000 (здесь примерно 5200 значений, проверила опытным путем)
* "**Высший**" - 30000000 <= x < 50000000 (здесь около 188 значений)
* "**Премиум**" - 50000000 <= x < 150000000 (здесь примерно 130 значений)
* "**Люкс**" - от 150000000 и до конца (здесь около 15 значений)

Захотелось немного детализировать группы с дорогими квартирами, хотя их, конечно, очень мало. Естественно, мы сосредоточимся вокруг значений, которые попадают в межквартильный размах. 

```python
#напишем функцию для категоризации данных

def price_segment(last_price):
    if 12190 <= last_price < 1700000:
        return 'Другое'
    if 1700000 <= last_price < 3400000:
        return 'Низший'
    if 3400000 <= last_price < 4650000:
        return 'Ниже среднего'
    if 4650000 <= last_price <= 6799000:
        return 'Средний'
    if 7000000 <= last_price < 30000000:
        return 'Выше среднего'
    if 30000000 <= last_price < 50000000:
        return 'Высший'
    if 50000000 <= last_price < 150000000:
        return 'Премиум'
    return 'Люкс'

estate_data['price_segment'] = estate_data['last_price'].apply(price_segment)

estate_data['price_segment'].value_counts()
```

Хочу посмотреть на взаимосвязь данных ***price_segment*** и ***days_exposition***, чтобы понять, как быстро продаются квартиры разных ценовых категорий.

```python
estate_data.groupby('price_segment').days_exposition.agg(['count','mean'])
```

```python
#посмотрим как сильно отклоняются средние значения по сегментам от среднего по выборке

display(f'Среднее значение по выборке {estate_data.days_exposition.mean()}')

t = estate_data.groupby('price_segment').days_exposition.agg(['count','mean'])

t['deviation_all_mean'] = (t['mean'] - estate_data.days_exposition.mean())/estate_data.days_exposition.mean()
t.style.format({
                'deviation_all_mean':'{:.2%}'
               })
```

Сегмент "средний", на мой взгляд, тоже имеет небольшое отклонение, сделки по квартирам стоимостью от 4,6 млн до 7 млн также продаются быстрее, чем остальные. Чтобы посмотреть на средние отклонения по группам, снова сделаем категоризацию:

* **"в центре"** - 208 <= x <= 7000
* **"рядом с центром"** - 7000 < x <= 10000
* **"недалеко от центра"** - 10000 < x <= 15000
* **"далеко от центра"** - 15000 < x <= 27000 
* **"очень далеко от центра"** - 27000 < x <= 35000
* **"область"** - x > 35000

```python
def cityCenter_group(distance):
    if 208 <= distance <= 7000:
        return 'в центре'
    if 7000 < distance <= 10000:
        return 'рядом с центром'
    if 10000 < distance <= 15000:
        return 'недалеко от центра'
    if 15000 < distance <= 27000:
        return 'далеко от центра'
    if 27000 < distance <= 35000:
        return 'очень далеко от центра'
    return 'область'

estate_data_filter['cityCenter_group'] = estate_data_filter['cityCenters_nearest'].apply(cityCenter_group)
estate_data_filter['cityCenter_group'].value_counts()
```

**Комментарий:**
<br>Такие границы были выбраны по моему мнению, безусловно при выполнении реального задания, нужно проконсультироваться с заказчиком.
